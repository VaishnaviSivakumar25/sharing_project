## Google Pay Expenses sharing

- Business Use cases: Friends want to easily track shared expenses and settle debts using a simple tool similar to Google Pay.
- Descption: Create a python progra that helps friends manage shared expenses and calculate who owes whom.  Suppose there are three friends---Alice, Bob and carol.  They went on a trip and shared expenses for accodation ans meals.  The program will allow them to input these expenses.  Calculate how much each person needs to pay or be reimburshed, and display the final settlement amounts.

## 1.Three person in the list as per the question given 

case: 1

```python
class ExpenseSharing:
    def __init__(self, friends):
        self.friends = friends
        self.expenses = {friend: 0 for friend in friends}

    def add_expense(self, payer, amount, participants):
        split_amount = amount / len(participants)
        for participant in participants:
            if participant != payer:
                self.expenses[participant] += split_amount
                self.expenses[payer] -= split_amount

    def calculate_settlement(self):
        for friend, balance in self.expenses.items():
            if balance > 0:
                print(f"{friend} needs to be reimburshed: Rs.{balance:.2f}")
            elif balance < 0:
                
                print(f"{friend} owes: Rs.{-balance:.2f}")
            else:
                print(f"{friend} is settled up.")
```

```python
if __name__ == "__main__":
    friends = ["Alice", "Bob", "Carol"]
    expense_sharing = ExpenseSharing(friends)

    #sample expenses
    expense_sharing.add_expense("Alice", 900, ["Alice", "Bob", "Carol"])
    expense_sharing.add_expense("Bob", 300, ["Alice", "Bob", "Carol"])
    expense_sharing.add_expense("Carol", 200, ["Alice", "Carol"])

    #calculate Settlement
    print("\n Final Settlement:")
    expense_sharing.calculate_settlement()
```

## Identifying who's the debitor and creditor

Case: 2

```python
class ExpenseSharing:
    def __init__(self, friends):
        self.friends = friends
        self.balances = {friend: 0 for friend in friends}

    def add_expense(self, payer, amount, participants):
        split_amount = amount / len(participants)
        for participant in participants:
             self.balances[participant] -= split_amount
        self.balances[payer] += amount

    def calculate_settlement(self):
        creditors = []
        debitors = []


        for friend, balance in self.balances.items():
            if balance > 0:
                creditors.append((friend, balance))
            elif balance < 0:
                debitors.append((friend, -balance))
        while debitors and creditors:
            debitor, debit_amount = debitors.pop()
            creditor, credit_amount = creditors.pop()

            payment = min(debit_amount, credit_amount)

            print(f"{debitor} owes {creditor}: Rs.{payment:.2f}")

            if debit_amount > payment:
                debitors.append((debitor, debit_amount - payment))
            if credit_amount > payment:
                creditors.append((creditor, credit_amount -payment))
            
            
```

```python
if __name__ == "__main__":
    friends = ["Alice", "Bob", "Carol"]
    expense_sharing = ExpenseSharing(friends)

    #sample expenses
    expense_sharing.add_expense("Alice", 900, ["Alice", "Bob", "Carol"]) # accodmation
    expense_sharing.add_expense("Bob", 300, ["Alice", "Bob", "Carol"]) # hotel
    expense_sharing.add_expense("Carol", 200, ["Alice", "Carol"]) #fuel

    #calculate Settlement
    print("\n Final Settlement:")
    expense_sharing.calculate_settlement()
```
